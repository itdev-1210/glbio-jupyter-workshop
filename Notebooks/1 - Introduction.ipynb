Now lets do some plots:  (note how command completion will help inside of the seq function; and how jupyter will match parens for you)

```r
time <- seq(from = 0, to=4*pi, length.out = 100)
intensity <- exp(sin(sin(time)))
```

```r
plot(time, intensity)
```

```r
df  <- data.frame(time, intensity)
```

```r
library(ggplot2)
```

```r
ggplot(df) + geom_point(aes(x=time, y=intensity))
```
